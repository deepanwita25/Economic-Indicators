```python
BASE_API = "https://www.imf.org/external/datamapper/api/v1/countries"
```

```python
r = requests.get(BASE_API)
```

```python
if r.status_code == 200:
    print("Success!")
elif r.status_code == 404:
    print("Error!!!")
```
