```python
import pandas as pd
import os
import requests
from pathlib import Path
```

# 1. Country code 3 digit from IMF

```python
BASE_API = "https://www.imf.org/external/datamapper/api/v1/countries"
```

```python
r = requests.get(BASE_API)
```

```python
if r.status_code == 200:
    print("Success!")
elif r.status_code == 404:
    print("Error!!!")
```

```python
r.headers['Content-Type']
```

```python
data_json = r.json()
countries = data_json['countries']
countries
```

```python
country_list = []
for country_code, label in countries.items():
    country_list.append({'code':country_code, 'country_name': label['label']})
```

```python
country_list
```

```python
countries_list = pd.DataFrame(country_list)
countries_list
```

```python
countries_list.isna().sum()
```

```python
countries_list[countries_list.isna().any(axis=1)]
```

```python
countries_list = countries_list[countries_list['country_name'].notna()]
```

```python
countries_list.isna().sum()
```

```python
countries_list=countries_list.drop_duplicates()
```

# 2. Country code 2 digit

```python
data_csv = "data_source/country_iso_2_digit.csv"
```

```python
path_csv = Path.cwd().parent / data_csv
path_csv
```

```python
iso_df = pd.read_csv(path_csv, encoding = 'utf-8')
iso_df
```

# 3. Combining 2 dfs

```python
df_iso2 = pd.merge(iso_df, countries_list, how = 'inner', left_on= 'Name', right_on= 'country_name')
df_iso2
```

```python
# Drop the extra country_name column
df_iso2 = df_iso2.drop(columns =['Name'], axis=1)
df_iso2
```

```python
df_iso2= df_iso2.rename(columns = {'Code':'iso_code_2', 'code':'iso_code_3'})
```

```python
df_iso2
```

```python
# Re-ordering columns
df_iso = df_iso2[['country_name', 'iso_code_2', 'iso_code_3']]
df_iso
```

```python
# Checking for Nans
df_iso.isna().sum()
```
